```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic function
def logistic_function(z):
    return 1 / (1 + np.exp(-z))

# Define the model
def log_odds(hours_studied, review_session):
    return -3 + 0.8 * hours_studied + 1.5 * review_session

# Generate some data
hours_studied = 2

# Calculate log-odds and probabilities for both Review_Session=0 and Review_Session=1
log_odds_0 = log_odds(hours_studied, 0)
probability_0 = logistic_function(log_odds_0)
print('Log_odds when revision session is 0 is=',log_odds_0)
print('odds when revision session is 0 is =',np.exp(log_odds_0))
print('the likelihood of passing the exam when revision session is 0 is=',probability_0)


log_odds_1 = log_odds(hours_studied, 1)
probability_1 = logistic_function(log_odds_1)
print('Log_odds when revision session is 1 =',log_odds_1)
print('odds when revision session is 1 is  =',np.exp(log_odds_1))
print('the likelihood of passing the exam when revision session is 1 is=',probability_1)
# Plot the data
plt.figure(figsize=(10,6))

plt.plot(hours_studied, probability_0, label='Review_Session=0')
plt.plot(hours_studied, probability_1, label='Review_Session=1', linestyle='--')

plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Effect of Hours Studied and Review Session on Passing Probability')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
